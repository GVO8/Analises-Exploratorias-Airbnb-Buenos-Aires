# Séries temporais e agregações

### Crie uma série temporal diária com o preço médio diário apenas para dias disponíveis \(available == 't'\)\. Armaze a série em um novo DataFrame com uma coluna chamada price\_mean\_day\.

```
df_disponivel = df[df['available'] == 't'].copy()

df_price_mean_day = df_disponivel.groupby('date')['price'].mean().reset_index()
df_price_mean_day = df_price_mean_day.rename(columns={'price': 'price_mean_day'})
df_price_mean_day = df_price_mean_day.sort_values('date').reset_index(drop=True)
df_price_mean_day
```

### Reamostre os dados para bases semanais e mensais utilizando resample\('W'\) e resample\('M'\)\.

```
df_price_mean_day['date'] = pd.to_datetime(df_price_mean_day['date'])
df_temp = df_price_mean_day.set_index('date')

df_semanal = df_temp.resample('W').agg({
    'price_mean_day': 'mean'
}).reset_index()

df_mensal = df_temp.resample('M').agg({
    'price_mean_day': 'mean'
}).reset_index()

df_semanal = df_semanal.rename(columns={'price_mean_day': 'price_mean_week'})
df_mensal = df_mensal.rename(columns={'price_mean_day': 'price_mean_month'})

print("Dados Semanais:")
print(df_semanal.head())

print("Dados Mensais:")
print(df_mensal.head())
```

### Plote gráficos de linha para as séries diária, semanal e mensal, destacando tendências e sazonalidades\.

```
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

def add_trendline(ax, x, y, color):
    if len(x) > 1:
        x_numeric = np.arange(len(x))
        z = np.polyfit(x_numeric, y, 1)
        p = np.poly1d(z)
        ax.plot(x, p(x_numeric), linestyle='--', color=color, 
                linewidth=2, alpha=0.8, label='Tendência')

axes[0].plot(df_price_mean_day['date'], df_price_mean_day['price_mean_day'], 
             linewidth=1, alpha=0.7, color='blue')
add_trendline(axes[0], df_price_mean_day['date'], df_price_mean_day['price_mean_day'], 'red')
axes[0].set_title('Série Diária com Tendência')
axes[0].set_ylabel('Preço')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(True, alpha=0.3)
axes[0].legend()

axes[1].plot(df_semanal['date'], df_semanal['price_mean_week'], 
             linewidth=2, marker='o', color='orange')
add_trendline(axes[1], df_semanal['date'], df_semanal['price_mean_week'], 'red')
axes[1].set_title('Série Semanal com Tendência')
axes[1].set_ylabel('Preço Semanal')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(True, alpha=0.3)
axes[1].legend()

axes[2].plot(df_mensal['date'], df_mensal['price_mean_month'], 
             linewidth=3, marker='s', color='green')
add_trendline(axes[2], df_mensal['date'], df_mensal['price_mean_month'], 'red')
axes[2].set_title('Série Mensal com Tendência')
axes[2].set_ylabel('Preço Mensal')
axes[2].tick_params(axis='x', rotation=45)
axes[2].grid(True, alpha=0.3)
axes[2].legend()

plt.tight_layout()
plt.show()
```

### Crie uma amostra estratificada por room\_type com proporções semelhantes às do dataset original\.

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

def amostra_estratificada_proporcional(df, coluna_strata, tamanho_amostra=0.1, random_state=42):
    proporcao = tamanho_amostra
    amostra, _ = train_test_split(
        df,
        test_size=1-proporcao,
        stratify=df[coluna_strata],
        random_state=random_state
    )
    
    return amostra

amostra_estratificada_proporcional(df, 'room_type', tamanho_amostra=0.2)
```

# Padronização e transformação de escala

### Aplique a transformação StandardScaler nas variáveis numéricas \(price, accommodates, minimum\_nights, availability\_365\)\.

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
colunas_numericas = ["price", "accommodates", "minimum_nights", "availability_365"]
df_sc = df[colunas_numericas]
infos_scaled = scaler.fit_transform(df_sc)
df_sc_scaled = pd.DataFrame(infos_scaled, columns=colunas_numericas, index=df_sc.index)
df_sc_scaled
```

### Mostre estatísticas antes e depois da padronização, verificando se as médias se aproximam de 0 e as variâncias de 1\.

```
import matplotlib.pyplot as plt
import seaborn as sns

pic, axes = plt.subplots(4, figsize=[10,8])
sns.lineplot(df.price, ax=axes[0])
sns.lineplot(df_sc_scaled.price, ax=axes[0])

m_price_df = df.price.mean()
m_price_df_sc = df_sc_scaled.price.mean()

axes[0].axhline(m_price_df, color='red', linestyle='--', label=f'Média price df: {m_price_df:.2f}')
axes[0].axhline(m_price_df_sc, color='green', linestyle='--', label=f'Média price scaled: {m_price_df_sc:.2f}')

sns.lineplot(df.accommodates, ax=axes[1])
sns.lineplot(df_sc_scaled.accommodates, ax=axes[1])

m_accommodates_df = df.accommodates.mean()
m_accommodates_df_sc = df_sc_scaled.accommodates.mean()

axes[1].axhline(m_accommodates_df, color='red', linestyle='--', label=f'Média accommodates df: {m_accommodates_df:.2f}')
axes[1].axhline(m_accommodates_df_sc, color='green', linestyle='--', label=f'Média accommodates scaled: {m_accommodates_df_sc:.2f}')

sns.lineplot(df.minimum_nights, ax=axes[2])
sns.lineplot(df_sc_scaled.minimum_nights, ax=axes[2])

m_minimum_nights_df = df.minimum_nights.mean()
m_minimum_nights_df_sc = df_sc_scaled.minimum_nights.mean()

axes[2].axhline(m_minimum_nights_df, color='red', linestyle='--', label=f'Média minimum_nights df: {m_minimum_nights_df:.2f}')
axes[2].axhline(m_minimum_nights_df_sc, color='green', linestyle='--', label=f'Média minimum_nights scaled: {m_minimum_nights_df_sc:.2f}')

sns.lineplot(df.availability_365, ax=axes[3])
sns.lineplot(df_sc_scaled.availability_365, ax=axes[3])

v_availability_365_df = df.availability_365.var()
v_availability_365_df_sc = df_sc_scaled.availability_365.var()

axes[3].axhline(v_availability_365_df, color='red', linestyle='--', label=f'Variância availability_365 df: {v_availability_365_df:.2f}')
axes[3].axhline(v_availability_365_df_sc, color='green', linestyle='--', label=f'Variância availability_365 scaled: {v_availability_365_df_sc:.2f}')
```

### Use inverse\_transform para reconverter os valores padronizados e valide numericamente o retorno à escala original\.

```
# como é possivel verificar no head, os valores são iguais, portanto na mesma escala

info_orig = scaler.inverse_transform(df_sc_scaled)

df_orig = pd.DataFrame(info_orig, columns=colunas_numericas, index=df_sc.index)

print(f"df original \n{df_sc.head()}\n")
print(f"df reconvertido \n{df_orig.head()}")
```

### Explique a importância da padronização para experimentos preditivos e comparações entre variáveis com magnitudes distintas\.

É imprecindivel a padronização de variáveis numéricas, pois muitos algoritmos, como o próprio de regressão linear, outras regressões, algoritmos de deepleaning, entre outros, são evetados por diferenças na escala desses valores, gerando assim um viés.

# Janelamento, suavização e previsão simples

### Calcule médias móveis de 7 e 30 dias para a série de preços \(price\_mean\_day\) usando rolling\(\)\.

```
# Como a coluna price_mean_day, não existe, vou utilizar a coluna price, proveniente do dataset de calendar.
df = df.sort_values('date')

df['price_mean_7d'] = df['price'].rolling(window=7, min_periods=1).mean()
df['price_mean_30d'] = df['price'].rolling(window=30, min_periods=1).mean()
df
```

### Calcule a variação percentual mês a mês \(pct\_change\(\)\) do preço médio mensal e destaque períodos de oscilação atípica\.

```
df['date'] = pd.to_datetime(df['date'])
df_monthly = df.resample('M', on='date')['price'].mean().reset_index()
df_monthly.columns = ['date', 'price_mean_monthly']

df_monthly['price_monthly_pct_change'] = df_monthly['price_mean_monthly'].pct_change() * 100

mean_change = df_monthly['price_monthly_pct_change'].mean()
std_change = df_monthly['price_monthly_pct_change'].std()

df_monthly['atypical_period'] = abs(df_monthly['price_monthly_pct_change'] - mean_change) > (2 * std_change)

df_monthly['atypical_streak'] = df_monthly['atypical_period'].astype(int)
df_monthly['atypical_streak_id'] = (df_monthly['atypical_period'] != df_monthly['atypical_period'].shift()).cumsum()

atypical_periods = df_monthly[df_monthly['atypical_period']]
print(f"Total de períodos atípicos: {len(atypical_periods)}")
print(f"Percentual de períodos atípicos: {(len(atypical_periods)/len(df_monthly))*100:.1f}%")

if len(atypical_periods) > 0:
    print("\nPeríodos com oscilações atípicas:")
    print(atypical_periods[['date', 'price_mean_monthly', 'price_monthly_pct_change']].to_string())
    
    print("\nEstatísticas das oscilações atípicas:")
    print(f"Maior alta atípica: {atypical_periods['price_monthly_pct_change'].max():.2f}%")
    print(f"Maior queda atípica: {atypical_periods['price_monthly_pct_change'].min():.2f}%")
    print(f"Média das oscilações atípicas: {atypical_periods['price_monthly_pct_change'].mean():.2f}%")
```

```
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

df_temp = df.set_index('date').sort_index()

alpha = 0.3
model = SimpleExpSmoothing(df_temp['price']).fit(smoothing_level=alpha, optimized=False)
df_temp['smoothed'] = model.fittedvalues

pic, ax = plt.subplots(2, figsize=(10,8))

sns.lineplot(df_temp['price'], 
             label='Série Bruta', color='blue', alpha=0.5, ax=ax[0])

sns.lineplot(df_temp['smoothed'], 
             label=f'Suavizada (α={alpha})', color='red', ax=ax[1])

ax[0].set_title('Comparação: Bruta vs Suavizada', fontsize=14)
ax[0].set_ylabel('Preço')
plt.legend()

plt.tight_layout()
plt.show()
```
